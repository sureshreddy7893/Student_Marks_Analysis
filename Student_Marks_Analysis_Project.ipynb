```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("student_marks_Dataset.xlsx")
print(df.head(5))
print(df.shape)
```

###Convert marks to array

```python
subjects=["Math","Science","English","History","Computer"]
marks_arr=df[subjects].to_numpy()
marks_arr
```

###Perform Numpy Analysis

```python
total_Marks=np.sum(marks_arr,axis=1)
average_Marks=np.mean(marks_arr,axis=1)
subjact_Max=np.max(marks_arr,axis=0)
subject_Min=np.min(marks_arr,axis=0)
subject_Avg=np.mean(marks_arr,axis=0)
df["Total"]=total_Marks
df["Average"]=average_Marks.round(2)
```

###Grade distribution

```python

plt.figure(figsize=(7,5))
df["Grade"].value_counts().sort_index().plot(kind="bar",color="skyblue")
plt.title("Grade Distribution")
plt.xlabel("Grade")
plt.ylabel("Number of Students")
plt.grid(axis='y')
plt.show()
```

###Top 5 and Bottum 5 performers

```python
print("Top 5 Performers:")
print(df.sort_values(by="Total",ascending=False)[["StudentID","Total","Grade"]].head(5))
print("Bottum 5 Performers:")
print(df.sort_values(by="Total")[["StudentID","Total","Grade"]].head(5))
```

###Gender vs Average Marks

```python
df.groupby("Gender")["Average"].mean().plot(kind="bar",title="Average Marks by Gender",color=["purple","orange"])
plt.xlabel("Gender")
plt.ylabel("Average Marks")
plt.show()
```

###Attendence impact on performence

```python
plt.scatter(df["Attendance (%)"],df["Average"],alpha=0.5)
plt.title("Attendence vs Average Marks")
plt.xlabel("Attendence (%)")
plt.ylabel("Average Marks")
plt.grid(True)
plt.show()
```

###Rank Students

```python
df["Rank"]=df["Total"].rank(ascending=False,method='min').astype(int)
df=df.sort_values('Rank')
df
```

###Add pass/Fail

```python
df["Result"]=np.where(df["Average"]>=35,"Pass","Fail")
df
```

###Subject_wise_Analysis

```python
for subject in subjects:
  print(f"{subject}-Avg:{df[subject].mean():.2f},Max:{df[subject].max()},Min:{df[subject].min()}")
```

###correlation between subjects

```python
plt.figure(figsize=(8,5))
sns.heatmap(df[subjects].corr(),annot=True,cmap="coolwarm")
plt.title("Subject correlation Heatmap")
plt.show()
```
